```python
# Load CSV to check column names
csv_file = "E:/DA_PROJECTS/BANKING_DATA_ANALYTICS/DATASETS/Banking.csv"
df = pd.read_csv(csv_file)
df.head()
```

```python
bins = [0,100000,300000,float('inf')]
labels = ['Low', 'Mid','High']

df['Income Band'] = pd.cut(df["Estimated Income"],bins=bins, labels=labels, right=False)
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

```python
# Examine the distribution of unique categories in categorical coloumns
categorical_cols = df[['BRId','GenderId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','IAId','Income Band']].columns

for col in categorical_cols:
    print(f"Value Counts for '{col}':")
    display(df[col].value_counts())
```

```python
for i, predictor in enumerate(df.columns):
    plt.figure(i)
    sns.countplot(data=df,x=predictor, hue='GenderId')
```

```python

for i, predictor in enumerate(df[categorical_cols]):
    plt.figure(i)
    sns.countplot(data=df,x=predictor)#, hue='GenderId')
```

```python
for i, predictor in enumerate(df[categorical_cols]):
    plt.figure(i)
    sns.countplot(data=df,x=predictor, hue='Nationality')
```

```python
# HistPlot of Value of Counts with Different Occupation

for col in categorical_cols:
    if col == "Occupation":
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.title("Histogram of Occupation Count.")
    plt.show()
```

Numerical Analysis

```python
numerical_columns = ["Estimated Income","Superannuation Savings","Credit Card Balance","Bank Loans","Bank Deposits","Checking Accounts","Saving Accounts","Foreign Currency Account","Business Lending"]

#Univariate Analysis
plt.figure(figsize=(30,30))
for i, col in enumerate(numerical_columns):
    plt.subplot(4,3,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
plt.show()
```

Heatmaps

```python
correlation_matrix = df[numerical_columns].corr()

plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix,annot=True,cmap='crest',fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```
